#### Задание 3
Напишите код на Python, реализующий построение графиков:
окружности, эллипса,гиперболы.

```python
x = []
y1 = []
y2 =[]
R = 4.0
k = -R
while k <= R:
    x.append(k)
    y1.append(math.sqrt(R**2 - k**2))
    y2.append(-math.sqrt(R**2 - k**2))
    k += 0.0001
plt.figure(figsize=(5, 5))
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='blue')
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title('Окружность')
plt.show()
```

```python
x = []
y1 = []
y2 =[]
a = 4.0
b = 2.0
k = -a
while k <= a:
    x.append(k)
    y1.append(math.sqrt(b**2 - (b**2*k**2)/a**2))
    y2.append(-math.sqrt(b**2 - (b**2*k**2)/a**2))
    k += 0.0001
plt.figure(figsize=(5, 5))
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='red')
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title('Эллипс')
plt.show()
```

```python
x1 = []
x2 = []
y1 = []
y2 = []
y3 = []
y4 = []
a = 2.0
b = 3.0
k = -7.0
while k < 7:
    if k <= -a: 
        x1.append(k)
        y1.append(math.sqrt((b**2*k**2/a**2) - b**2))
        y2.append(-math.sqrt((b**2*k**2/a**2) - b**2))
    if k >= a: 
        x2.append(k)
        y3.append(math.sqrt((b**2*k**2/a**2) - b**2))
        y4.append(-math.sqrt((b**2*k**2/a**2) - b**2))
    k += 0.0001
plt.figure(figsize=(5, 5))
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(x1, y1, color='green')
plt.plot(x1, y2, color='green')
plt.plot(x2, y3, color='green')
plt.plot(x2, y4, color='green')
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title('Гипербола')
plt.show()
```

#### Задание 5, задача 1
Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np. meshgrid(X, Y)
Z = 2*X + Y
ax.plot_surface(X, Y, Z)
Z = 2*X + Y + 20
ax.plot_surface(X, Y, Z)
plt.show()
```

#### Задание 5, задача 2
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = plt.figure(figsize = (15,8))
ax1 = fig.add_subplot(1,2,1, projection = '3d')
ax2 = fig.add_subplot(1,2,2,projection = '3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2 + Y**2)
ax1.plot_wireframe(X, Y, Z)
ax1.set_title('Конус')
Z = X**2 - Y**2
ax2.plot_wireframe(X, Y, Z, color='green')
ax2.set_title('Гиперболический параболоид')
plt.show()
```
