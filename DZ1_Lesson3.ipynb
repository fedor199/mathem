#### Задание 3, задача 1
Напишите код, который будет переводить полярные координаты в декартовы.

```python
R = 4
alpha = np.pi/3
x = R * np.cos(alpha)
y = R * np.sin(alpha)
plt.scatter(x, y, color='red')
print("x = {:.2f}, y = {:.2f}".format(x, y))
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

#### Задание 3, задача 2
Напишите код, который будет рисовать график окружности в полярных координатах.

```python
x = []
y = []
R = 6.0
k = -R
while k <= R:
    x.append(k)
    y.append(R)
    k += 0.0001
plt.polar(x, y, color='blue')
plt.show()
```

#### Задание 4, пример 1
Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,7)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, (y - (np.exp(x) + x - 1)/x))

x1, y1 = fsolve(equations, (-1, 1))

print (x1, y1)
```

Решите систему уравнений и неравенств:

#### Область решения системы выделена синим цветом на графике

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax

x = np.linspace(-3, 0, 201)
ax1.plot(x, (np.exp(x) + x - 1)/x)
ax1.plot(x, x**2 - 1)
#ax1.xlabel('x')
#ax1.ylabel('y')
ax1.grid(True)
ax1.fill_between(x, (np.exp(x) + x - 1)/x, 8)

x = np.linspace(0, 3, 201)
ax2.plot(x, (np.exp(x) + x - 1)/x)
ax2.plot(x, x**2 - 1)
ax2.grid(True)
ax2.fill_between(x, (np.exp(x) + x - 1)/x, -1)
fig.set_size_inches(12, 6)
```
